```python
# I - importing library
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
# II - importing dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
# Print the keys:
cancer.keys()
```

```python
print(cancer.target_names)
```

```python
# Overview of feature names for the dataset:
print(cancer.feature_names)
```

```python
df_cancer = pd.DataFrame(cancer.data)
```

```python
df_cancer.head()
```

```python
df_cancer.columns = cancer.feature_names
```

```python
df_cancer.head()
```

```python
#Add new column
df_cancer['target'] = cancer.target
```

```python
df_cancer.head()
```

```python
# III - Visualising the data
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
# Heat map - displays correlation between all categories:
plt.figure(figsize=(20,10))
sns.heatmap(df_cancer.corr(), cmap='winter', annot=True)
plt.show()
```

```python
#Model Training
X = df_cancer.drop(['target'], axis = 1)

# this is the target:
y = df_cancer['target']
```

```python
#Splitting, tsting and training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
#Training model on NaiveBayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
#Evaluating the model
y_predict = nb.predict(X_test)
```

```python
#Making of confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,y_predict))
```
